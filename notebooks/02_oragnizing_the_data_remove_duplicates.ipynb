```python
import json 
import os
import pandas as pd
```

```python
dir = "../data"
all_json_files = os.listdir(dir)

all_dfs =[]
for file in all_json_files:
    if ".json" in file:
        with open("../data/"+file, 'r') as f:
            json_obj = json.loads(f.read())
            if not "service-error" in json_obj.keys() and not "error-response" in json_obj.keys():
                try:
                    df = pd.DataFrame(json_obj["search-results"]["entry"])
                    all_dfs.append(df)
                except Exception as e:
                    print(str(e), json_obj)
    

all_dfs_concat = pd.concat(all_dfs)
all_dfs_concat
```

```python
# all_dfs_concat = all_dfs_concat.drop_duplicates()
all_dfs_concat = all_dfs_concat.reset_index(drop=True)
```

```python
all_dfs_concat = all_dfs_concat.drop_duplicates("dc:identifier")
```

```python
all_dfs_concat = all_dfs_concat.reset_index(drop=True)
all_dfs_concat.to_csv("02_papers_with_abstracts_removed_duplicates_not_filtered_yet.csv", index=False)
```
