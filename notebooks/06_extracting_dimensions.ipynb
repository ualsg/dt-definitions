```python

def getKeywords (file):
    with open ("."+file, 'r') as f:
        text = f.read()
    # regex_pattern = r'\bKeywords\b(.*?)\bIntroduction\b'
    regex_pattern = r'\bKeywords\b\s+((?:\b\w+\b\s+){1,30})\bIntroduction\b'
    matches = re.search(regex_pattern, text, re.DOTALL)

    if matches:
        extracted_words = matches.group(1).strip()
        return extracted_words
    else:
        return ""
    

def getAbstract (file):
    with open ("."+file, 'r') as f:
        d = json.loads(f.read())
    try:
        return d["full-text-retrieval-response"]['coredata']['dc:description']
    except:
        return ""
```

```python
df["keyworkds"] = df["file"].apply(lambda x : getKeywords(x))
df["abstract"] = df["file"].apply(lambda x : getAbstract(x))
```
