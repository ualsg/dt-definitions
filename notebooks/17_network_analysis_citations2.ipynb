```python
import re

citation_list = [str(i) for i in list(df["citation_formatted"]) ]

new_list = []
for citation in citation_list:
    match = re.match(r"(\w+).*\((\d{4})\)", citation)
    if match:
        new_list.append(match.group(1) + " (" + match.group(2) + ")")
    else:
        match = re.match(r"(\w+).*\b(\d{4})\b", citation)
        if match:
            new_list.append(match.group(1) + " (" + match.group(2) + ")")
        else:
            new_list.append(citation)

print(new_list)
```

```python
df["citation_formatted_shortened"] = new_list
```

```python
matches = {}
```

```python
from fuzzywuzzy import process

# get the list of names
names = list(df["citation_formatted_shortened"])[:]

# create an empty dictionary to store the matches

# iterate over each name in the list
for i, name in enumerate(names):
    try:
        print(i, end="\r")
        # find the top 5 matches for each name
        top_matches = process.extract(name, names, limit=50)
        
        # add the matches to the dictionary
        matches[name] = top_matches
    except:
        pass

matches
```

```python
filtered_matches = {k: [pair for pair in v if pair[1] > 92] for k, v in matches.items()}
```

```python
filtered_matches2 = {k : [i[0] for i in v] for k, v in filtered_matches.items()}
```

```python
t = ""
for k, v in filtered_matches2.items():
    t+= k + "\n"
    t += str(v) + "\n\n\n"


with open("matches.txt", "w") as f:
    f.write(t)
```

```python

def get_authors(x):
    try:
        with open(x["file"], 'r') as f:
            j_file = json.loads(f.read())["full-text-retrieval-response"]
        core_data = j_file["coredata"]
        try:
            return core_data["dc:creator"][0]["$"].lower().split(",")[0] + "_" + core_data["prism:coverDate"].split("-")[0]
        except:
            return core_data["dc:creator"]["$"].lower().split(",")[0]+ "_" + core_data["prism:coverDate"].split("-")[0]
    except:
        print(x["file"])
        return "error"

df["authors"] = df.apply(lambda x: get_authors(x), axis=1)

    
```
