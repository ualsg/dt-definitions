```python
import pandas as pd
import os
import json
import random
```

```python
import os

def list_files(startpath):
    for root, dirs, files in os.walk(startpath):
        for file in files:
            yield os.path.join(root, file)

# Example usage:
            
all_files = []
startpath = '../full_texts/'
for filepath in list_files(startpath):
    if ".json" in filepath:
        all_files.append(filepath)
```

```python
all_data = []
for file in all_files:
    with open(file, 'r') as f:
        ret = json.loads(f.read())["full-text-retrieval-response"]
        jfile = ret["coredata"]
        jfile["file"] = file
        jfile["full_text"] = ret["originalText"]
        try:
            jfile["refs"] = ret["originalText"].lower().split(". references")[-1]
        except:
            jfile["refs"] = ""
        all_data.append(jfile)
```

```python
df = pd.DataFrame(all_data)
df["prism:coverDate"] = pd.to_datetime(df['prism:coverDate'])
df.columns
```

```python
df_2 = df[['file','eid', 'dc:description', 'prism:coverDate',
       'prism:aggregationType', 'prism:url',
       'dc:creator', 'link',  'openaccessType',
       'pii', 'prism:publisher', 'dc:title', 'openaccess',
       'dcterms:isPartOf', 'dcterms:subject', 'openaccessArticle',
       'prism:publicationName', 'pubType', 'prism:coverDisplayDate', 'prism:doi',
        'dc:identifier','prism:issn', 'prism:issueName', 'articleNumber', 'prism:edition']]
```

```python
df_2.rename({
    'dc:description': 'abstract', 
    'prism:coverDate': 'date',
    'prism:aggregationType': "publication_type",
    'prism:url': 'url',
    'dc:creator': 'authors',
    'dc:title':'title',
    'dcterms:subject': 'subjects'
}, axis=1, inplace=True)

df_2
```

```python
def get_authors(x):
    try:
        return [i["$"] for i in x]
    except:
        try:
            return x["$"]
        except:
            return []
        
def get_link(x):
    try:
        return [i["@href"] for i in x]
    except:
        return []
    
def get_subject(x):
    try:
        return [i["$"] for i in x]
    except:
        return x
df_2["authors"] = df_2["authors"].apply(lambda x : get_authors(x))
df_2["link"] = df_2["link"].apply(lambda x : get_link(x))
df_2["subjects"] = df_2["subjects"].apply(lambda x : get_subject(x))
df_2
```

```python
df_2.to_csv("../datasets/finalised/00_clean_all_papers_details.csv", index=False)
```
